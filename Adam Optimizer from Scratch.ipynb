## 3. Adam Optimizer

<div style='background:#0d1628;border-left:3px solid #10b981;border-radius:0 10px 10px 0;padding:20px 24px;font-family:monospace;max-width:820px;margin-bottom:12px;'>
  <div style='font-size:11px;letter-spacing:0.12em;text-transform:uppercase;color:#10b981;margin-bottom:10px;'>// Section 3 — Adam Optimizer</div>
  <p style='color:#94a3b8;font-size:13px;line-height:1.6;margin:0 0 14px;'>Adam maintains two running averages per parameter: a first moment (mean of gradients) and a second moment (uncentred variance). Both are bias-corrected at each step. The update rule is:</p>
  <div style='background:#0a1020;border:1px solid #1a2540;border-radius:6px;padding:14px 18px;margin-bottom:14px;'>
    <code style='color:#7dd3fc;font-size:12px;line-height:2;'>
      m_t  = beta1 * m_{t-1}  + (1 - beta1) * g_t<br>
      v_t  = beta2 * v_{t-1}  + (1 - beta2) * g_t^2<br>
      m_hat = m_t  / (1 - beta1^t)<br>
      v_hat = v_t  / (1 - beta2^t)<br>
      theta = theta - lr * m_hat / (sqrt(v_hat) + eps)
    </code>
  </div>
  <p style='color:#94a3b8;font-size:13px;line-height:1.6;margin:0;'>The bias correction terms <code style='background:#0a1020;padding:2px 5px;border-radius:3px;color:#7dd3fc;'>(1 - beta^t)</code> compensate for the zero-initialisation of the moment vectors. Without them, early updates are systematically too small.</p>
</div>

```python
class Adam:
    """
    Adaptive Moment Estimation (Adam) optimizer.

    Reference: Kingma & Ba (2015) — https://arxiv.org/abs/1412.6980

    Parameters
    ----------
    lr    : learning rate (alpha)
    beta1 : 1st-moment decay factor
    beta2 : 2nd-moment decay factor
    eps   : numerical stability term
    """

    def __init__(self, lr=0.001, beta1=0.9, beta2=0.999, eps=1e-8):
        self.lr    = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.eps   = eps
        self.m = {}  # 1st-moment vectors
        self.v = {}  # 2nd-moment vectors

    def update(self, param_id, theta, grad, t):
        """
        In-place Adam update for a single parameter array.

        Parameters
        ----------
        param_id : unique string key, e.g. 'W1', 'b2'
        theta    : parameter array, modified in-place
        grad     : gradient array, same shape as theta
        t        : current training iteration (1-indexed)
        """
        # Initialise moment vectors on first encounter
        if param_id not in self.m:
            self.m[param_id] = np.zeros_like(theta)
            self.v[param_id] = np.zeros_like(theta)

        b1, b2 = self.beta1, self.beta2

        # Biased first and second moment estimates
        self.m[param_id] = b1 * self.m[param_id] + (1 - b1) * grad
        self.v[param_id] = b2 * self.v[param_id] + (1 - b2) * (grad ** 2)

        # Bias-corrected estimates
        m_hat = self.m[param_id] / (1 - b1 ** t)
        v_hat = self.v[param_id] / (1 - b2 ** t)

        # Parameter update
        theta -= self.lr * m_hat / (np.sqrt(v_hat) + self.eps)

    def reset(self):
        """Clear all moment state."""
        self.m.clear()
        self.v.clear()


print("Adam class defined.")
print(f"Default hyperparams: lr=0.001, beta1=0.9, beta2=0.999, eps=1e-8")
```

## 4. Neural Network

<div style='background:#0d1628;border-left:3px solid #a78bfa;border-radius:0 10px 10px 0;padding:20px 24px;font-family:monospace;max-width:820px;margin-bottom:12px;'>
  <div style='font-size:11px;letter-spacing:0.12em;text-transform:uppercase;color:#a78bfa;margin-bottom:10px;'>// Section 4 — Neural Network</div>
  <p style='color:#94a3b8;font-size:13px;line-height:1.6;margin:0 0 14px;'>A two-layer fully connected network. All forward and backward operations are written explicitly, making every gradient step transparent.</p>
  <ul style='color:#94a3b8;font-size:13px;line-height:1.9;margin:0;padding-left:18px;'>
    <li><strong style='color:#e2e8f0;'>Architecture.</strong> Input (2) → Dense → ReLU → Dense → Sigmoid → Output (1).</li>
    <li><strong style='color:#e2e8f0;'>Weight init.</strong> Standard normal scaled by 0.5. Biases initialised to zero.</li>
    <li><strong style='color:#e2e8f0;'>Loss.</strong> Binary cross-entropy: <code style='background:#0a1020;padding:2px 5px;border-radius:3px;color:#7dd3fc;'>-mean(y*log(a2) + (1-y)*log(1-a2))</code>.</li>
    <li><strong style='color:#e2e8f0;'>Backward pass.</strong> Chain rule applied layer by layer. Sigmoid+BCE derivative simplifies to <code style='background:#0a1020;padding:2px 5px;border-radius:3px;color:#7dd3fc;'>(a2 - y) / n</code>.</li>
  </ul>
</div>

```python
# Activation functions

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-np.clip(x, -500, 500)))

def relu(x):
    return np.maximum(0.0, x)

def relu_grad(x):
    return (x > 0).astype(float)


class TwoLayerNet:
    """
    Two-layer network: X -> [Dense + ReLU] -> [Dense + Sigmoid] -> y_hat
    Loss: binary cross-entropy.
    """

    def __init__(self, n_in=2, n_hidden=8, seed=0):
        rng = np.random.default_rng(seed)
        scale = 0.5
        self.W1 = rng.standard_normal((n_in,     n_hidden)) * scale
        self.b1 = np.zeros((1, n_hidden))
        self.W2 = rng.standard_normal((n_hidden, 1))        * scale
        self.b2 = np.zeros((1, 1))

    def forward(self, X):
        self._X  = X
        self._z1 = X @ self.W1 + self.b1
        self._a1 = relu(self._z1)
        self._z2 = self._a1 @ self.W2 + self.b2
        self._a2 = sigmoid(self._z2)
        return self._a2

    def loss(self, y):
        a2 = np.clip(self._a2, 1e-12, 1 - 1e-12)
        return float(-np.mean(y * np.log(a2) + (1 - y) * np.log(1 - a2)))

    def backward(self, y):
        n   = y.shape[0]
        # Output layer — sigmoid + BCE derivative simplifies to (a2 - y) / n
        dz2 = (self._a2 - y) / n
        dW2 = self._a1.T @ dz2
        db2 = dz2.sum(axis=0, keepdims=True)
        # Hidden layer
        da1 = dz2 @ self.W2.T
        dz1 = da1 * relu_grad(self._z1)
        dW1 = self._X.T @ dz1
        db1 = dz1.sum(axis=0, keepdims=True)
        return {"W1": dW1, "b1": db1, "W2": dW2, "b2": db2}

    def predict(self, X):
        return (self.forward(X) >= 0.5).astype(int)

    def accuracy(self, X, y):
        return float((self.predict(X) == y).mean() * 100)


# Sanity check — forward pass with random weights
net_test = TwoLayerNet(seed=42)
out = net_test.forward(X)
print(f"Forward pass output shape : {out.shape}")
print(f"Output values (untrained) : {out.flatten().round(4)}")
```

## 5. Training and Benchmark

<div style='background:#0d1628;border-left:3px solid #f43f5e;border-radius:0 10px 10px 0;padding:20px 24px;font-family:monospace;max-width:820px;margin-bottom:12px;'>
  <div style='font-size:11px;letter-spacing:0.12em;text-transform:uppercase;color:#f43f5e;margin-bottom:10px;'>// Section 5 — Training and Benchmark</div>
  <p style='color:#94a3b8;font-size:13px;line-height:1.6;margin:0 0 14px;'>Three optimizers are trained on identical networks with the same random seed. Adam uses lr=0.01. SGD uses lr=0.1. Momentum SGD uses lr=0.01 with a momentum coefficient of 0.9. Each runs for 3000 epochs.</p>
  <ul style='color:#94a3b8;font-size:13px;line-height:1.9;margin:0;padding-left:18px;'>
    <li><strong style='color:#e2e8f0;'>Adam.</strong> Adaptive per-parameter learning rates. Bias correction applied at every step.</li>
    <li><strong style='color:#e2e8f0;'>SGD.</strong> Plain gradient descent. Single global learning rate. No memory.</li>
    <li><strong style='color:#e2e8f0;'>Momentum.</strong> Accumulates a velocity vector. Smooths noisy gradients but no adaptive scaling.</li>
  </ul>
</div>

```python
def train(optimizer="adam", epochs=3000, lr=0.01, n_hidden=8, seed=0):
    """
    Train a TwoLayerNet on XOR.

    Returns (model, history) where history is a list of dicts
    with keys: epoch, loss, acc — recorded every 100 steps.
    """
    net = TwoLayerNet(n_in=2, n_hidden=n_hidden, seed=seed)
    opt = Adam(lr=lr) if optimizer == "adam" else None

    vel = {k: np.zeros_like(p)
           for k, p in [("W1", net.W1), ("b1", net.b1),
                        ("W2", net.W2), ("b2", net.b2)]}

    history = []

    for t in range(1, epochs + 1):
        net.forward(X)
        loss  = net.loss(y)
        grads = net.backward(y)

        for name, param in [("W1", net.W1), ("b1", net.b1),
                             ("W2", net.W2), ("b2", net.b2)]:
            g = grads[name]
            if optimizer == "adam":
                opt.update(name, param, g, t)
            elif optimizer == "sgd":
                param -= lr * g
            elif optimizer == "momentum":
                vel[name] = 0.9 * vel[name] + lr * g
                param    -= vel[name]

        if t % 100 == 0 or t == 1:
            history.append({
                "epoch": t,
                "loss":  round(loss, 6),
                "acc":   round(net.accuracy(X, y), 1),
            })

    return net, history
```

```python
# Run all three optimizers
EPOCHS = 3000

net_adam, hist_adam = train(optimizer="adam",     epochs=EPOCHS, lr=0.01)
net_sgd,  hist_sgd  = train(optimizer="sgd",      epochs=EPOCHS, lr=0.1)
net_mom,  hist_mom  = train(optimizer="momentum", epochs=EPOCHS, lr=0.01)

# Summary table
print(f"{'Optimizer':<20} {'Final Loss':<16} {'Accuracy'}")
print("-" * 50)
for label, hist in [("Adam (lr=0.01)",     hist_adam),
                    ("SGD  (lr=0.10)",     hist_sgd),
                    ("Momentum (lr=0.01)", hist_mom)]:
    row = hist[-1]
    print(f"{label:<20} {row['loss']:<16} {row['acc']}%")
```

```python
# Adam convergence curve
print(f"{'Epoch':<9} {'Loss':<14} {'Accuracy'}")
print("-" * 36)
for row in hist_adam:
    bar = "#" * int(row["acc"] / 10)
    print(f"{row['epoch']:<9} {row['loss']:<14} {row['acc']:5.1f}%  {bar}")
```

```python
# Final predictions
labels = ["[0,0]", "[0,1]", "[1,0]", "[1,1]"]
preds  = net_adam.forward(X).flatten()
hard   = net_adam.predict(X).flatten()

print("Adam — final predictions on XOR:")
print(f"{'Input':<10} {'Target':<10} {'Prob':<10} {'Predicted'}")
print("-" * 42)
for i in range(4):
    print(f"{labels[i]:<10} {int(y[i,0]):<10} {preds[i]:.5f}   {hard[i]}")
```
